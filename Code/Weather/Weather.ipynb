```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from scipy.stats import normaltest
from sklearn import datasets, preprocessing, metrics
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_excel('2012.xls', index_col=None)
df.dropna()
```

```python
df = df.rename(columns={"Maksimumtemperatur": "Maksimumtemperatur2012"})
```

```python
df1 = pd.read_excel('2013.xls', index_col=None)
df1.dropna()
```

```python
df1 = df1.rename(columns={"Maksimumtemperatur": "Maksimumtemperatur2013"})
```

```python
df2 = pd.read_excel('2014.xls', index_col=None)
df2.dropna()
```

```python
df2 = df2.rename(columns={"Maksimumtemperatur": "Maksimumtemperatur2014"})
```

```python
dfresult = pd.concat([df, df1, df2], axis=1)
```

```python
train_data = final.sample(frac=0.8, random_state=1)
test_data = final.drop(train_data.index)

# Definere mine input variable og output
X_train = train_data[['Maksimumtemperatur2012', 'Maksimumtemperatur2013']]
y_train = train_data['Maksimumtemperatur2014']
X_test = test_data[['Maksimumtemperatur2012', 'Maksimumtemperatur2013']]
y_test = test_data['Maksimumtemperatur2014']

# Laver en linear regression model
model = LinearRegression()

# Træner en med trænings data
model.fit(X_train, y_train)

# Tester modellen på testing dataen
y_pred = model.predict(X_test)

# Evaluaere performance på modellen ved brug af "mean squared error"
mse = mean_squared_error(y_test, y_pred)
print('Mean squared error:', mse)

# Bruger den trænede model til at predicte maximum temperaturen for 2015
X_new = pd.DataFrame({'Maksimumtemperatur2012': ['19'], 'Maksimumtemperatur2013': ['17']})
y_new = model.predict(X_new)
print('Predicted maximum temperature for 2015:', y_new)
```

#### in this example i am tpredicting the temperature in Celsius degrees and my MSE is 2.19, that means my model's predictions on average deviate by about 1.48 degrees Celsius from the true values.
