### For the start i want to use my text from my SP3,where i have been removing all the stop words, and did lemmatize tokens, and stem tokens. From this we can make a worldcloud, that will represent the words that is being used the most therefor these words can have a saying in what this text is about or what the essential about it is.

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

with open('kmd.txt', 'r') as f:
    text = f.read()
    
wordcloud = WordCloud(width=800, height=800, background_color='white').generate(text)

plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis('off')
plt.tight_layout(pad=0)

plt.show()
```

### Next up is two plots with the salaray comparison for the companies, to be able to visulaize the differences

```python
df = pd.read_excel('companies.xlsx')
```

```python
import matplotlib.pyplot as plt
import pandas as pd

salaries = df[[2020, 2021, 2022]]

years = [2020, 2021, 2022]

salaries.T.plot(kind='line')
plt.ylim(0, 10000)
plt.autoscale()
plt.legend(loc='best')

plt.title('Salary Comparison for Different Companies')
plt.xlabel('Year')
plt.ylabel('Salary')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

salaries = df[['company', 2020, 2021, 2022]]

salaries = salaries.set_index('company')

years = [2020, 2021, 2022]

salaries.T.plot(kind='line')

plt.legend(loc='best')

plt.title('Salary Comparison for Different Companies')
plt.xlabel('Year')
plt.ylabel('Salary')

plt.show()
```
