```python
import spacy
#!python -m spacy download da_core_news_sm
from spacy.lang.da.examples import sentences 

text = "En af den danske naturs allermest eftertragtede urter begynder lige så langsomt at dukke op i skovbunden. Men skal du ud at samle ramsløg, skal du være opmærksom, skriver Fødevarestyrelsen i en pressemeddelelse.Ofte side om side med ramsløgene vokser nemlig den næsten identiske liljekonval, som kan give alvorlige forgiftninger. Sidste år fik Giftlinjen en håndfuld opkald fra danskere, som havde symptomer på forgiftning efter forveksling, Især før liljekonvallen har sat sine hvide, klokkeformede blomster, som er nemme at skelne fra ramsløgens stjerneformede blomst, er der stor risiko for forveksling. For man kan nemt komme til at tage fejl af de grønne, aflange blade, som ligner hinanden, lyder det i pressemeddelelsen. Vil man være sikker på, at man kommer den rette plante i maden, kan man nulre bladene, Ramsløg har en kraftig hvidløgsagtig duft, mens liljekonval ingen har."

nlp = spacy.load("da_core_news_sm")
doc = nlp(text)
print(doc.text)
print("\n")
for token in doc:
    print(token.text, token.pos_, token.dep_)
```
